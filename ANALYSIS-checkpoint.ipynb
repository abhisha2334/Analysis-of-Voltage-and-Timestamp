```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Sample_Data.csv")
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'], dayfirst=True)
```

```python
df = df.sort_values('Timestamp').reset_index(drop=True)
```

```python
print("Data loaded successfully")
print(df.head())
```

```python
#Plot Voltage vs Timestamp
plt.figure(figsize=(12,5))
plt.plot(df['Timestamp'], df['Values'], linewidth=1)
plt.xlabel("Timestamp")
plt.ylabel("Voltage")
plt.title("Voltage vs Timestamp")
plt.xticks(rotation=45)
plt.ylim(20, 105)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_daily = df.set_index('Timestamp')
```

```python
## 5-day moving average
daily_voltage = df_daily['Values'].resample('D').mean()
```

```python
DA_5day = daily_voltage.rolling(window=5).mean()
```

```python
plt.figure(figsize=(12,5))
plt.plot(daily_voltage.index, daily_voltage, label="Daily Average Voltage", linewidth=1)
plt.plot(DA_5day.index, DA_5day, label="5-Day Moving Average", linewidth=3)
plt.xlabel("Date")
plt.ylabel("Voltage")
plt.title("Voltage vs Time with 5-Day Moving Average")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Find Local Peaks and Lows
from scipy.signal import find_peaks
```

```python
peaks, _ = find_peaks(df['Values'], distance=10)
```

```python
lows, _ = find_peaks(-df['Values'], distance=10)
```

```python
peaks_df = df.iloc[peaks][['Timestamp', 'Values']]
peaks_df['Type'] = 'Peak'
```

```python
lows_df = df.iloc[lows][['Timestamp', 'Values']]
lows_df['Type'] = 'Low'
```

```python
peaks_lows_table = pd.concat([peaks_df, lows_df]) \
                       .sort_values('Timestamp') \
                       .reset_index(drop=True)
```

```python
print("\nLocal Peaks and Lows:")
print(peaks_lows_table.head())
```

```python
plt.figure(figsize=(12,5))
plt.plot(df['Timestamp'], df['Values'], label='Voltage', linewidth=1, alpha=0.7)

plt.scatter(df.iloc[peaks]['Timestamp'],
            df.iloc[peaks]['Values'],
            color='red', label='Peaks', s=25)

plt.scatter(df.iloc[lows]['Timestamp'],
            df.iloc[lows]['Values'],
            color='green', label='Lows', s=25)

plt.xlabel("Timestamp")
plt.ylabel("Voltage")
plt.title("Voltage vs Timestamp with Local Peaks and Lows")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Find Voltage Below 20
below_20_df = df[df['Values'] < 20]
```

```python
below_20_table = below_20_df[['Timestamp', 'Values']] \
                    .rename(columns={'Values': 'Voltage'})
```

```python
print("\nVoltage Below 20:")
print(below_20_table)
print("Count below 20:", len(below_20_table))
```

```python
#Downward slope acceleration detection and print the timestamp 
df['slope'] = df['Values'].diff()
```

```python
df['acceleration'] = df['slope'].diff()
```

```python
downward_accel_df = df[
    (df['slope'] < 0) &
    (df['acceleration'] < 0)
]
```

```python
downward_accel_table = downward_accel_df[
    ['Timestamp', 'Values', 'slope', 'acceleration']
]
```

```python
print("\nDownward slope acceleration points:")
print(downward_accel_table.head())
```

```python
print("\nTimestamps where downward slope accelerates:")
print(downward_accel_table['Timestamp'])
```
